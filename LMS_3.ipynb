```python
#import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

```

```python
# Map target values to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

```

```python
# Exploratory Data Analysis (EDA)
print(df.head())


```

```python
print(df.info())
```

```python
print(df.describe())
```

### Data Visualization

```python
# Pairplot for pairwise relationships between features
sns.pairplot(df, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()

```

```python
# Boxplot for feature distribution by species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title('Boxplot of Petal Length by Species')
plt.show()
```

```python
# Violin plot for feature distribution by species
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='petal width (cm)', data=df)
plt.title('Violin plot of Petal Width by Species')
plt.show()
```

```python
# Scatter plot for Sepal length vs Sepal width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.show()
```

### plotting different pandas inbuilt visualization techniques


```python
# Plotting area plot
df.plot.area()
plt.title('Area Plot of Iris Dataset')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting horizontal bar plot
df.plot.barh()
plt.title('Horizontal Bar Plot of Iris Dataset')
plt.xlabel('Values')
plt.ylabel('Index')
plt.show()
```

```python
# Plotting density plot
df.plot.density()
plt.title('Density Plot of Iris Dataset')
plt.xlabel('Values')
plt.show()
```

```python
# Plotting histogram
df.plot.hist(alpha=0.5)
plt.title('Histogram of Iris Dataset')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting line plot
df.plot.line()
plt.title('Line Plot of Iris Dataset')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting bar plot
df['species'].value_counts().plot.bar()
plt.title('Bar Plot of Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting box plot
sns.boxplot(data=df.drop(columns='species'))
plt.title('Box Plot of Iris Dataset')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
# Plotting hexbin plot
df.plot.hexbin(x='sepal length (cm)', y='sepal width (cm)', gridsize=15)
plt.title('Hexbin Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Plotting KDE plot
sns.kdeplot(data=df['sepal length (cm)'], shade=True)
plt.title('Kernel Density Estimation (KDE) Plot of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
# Plotting pie chart
df['species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart of Species Distribution')
plt.ylabel('')
plt.show()
```
